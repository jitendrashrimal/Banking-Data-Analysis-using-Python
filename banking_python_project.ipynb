```python
from dateutil.relativedelta import relativedelta
from datetime import datetime
```

```python
def cal_years(value):
    end_date = datetime.today()
    difference = relativedelta(end_date, value)
    return difference.years
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
df["Joined Years"] = df["Joined Bank"].apply(cal_years)
df.head(1)
```

```python
select_col = df.loc[:, ['Fees', 'Business Lending']]
corr = select_col.corr().iloc[0, 1]
print(f"Correlation between 'Fees' and 'Business_Lending': {corr}")

plt.figure(figsize=(6, 4))
sns.heatmap(select_col.corr(), annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation between Fees and Business Lending')
plt.show()
```
